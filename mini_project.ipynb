**TWO TYPES OF ENCODING CATEGORICAL TECHNIQUES**

*   ONE HOT ENCODED
*   LABEL ENCODER
Label Encoding

Label Encoding is a popular encoding technique for handling categorical variables. In this technique, each label is assigned a unique integer based on alphabetical ordering.

One-Hot Encoding
One-Hot Encoding is another popular technique for treating categorical variables. It simply creates additional features based on the number of unique values in the categorical feature. Every unique value in the category will be added as a feature.
One-Hot Encoding is the process of creating dummy variables.
In this encoding technique, each category is represented as a one-hot vector.

**WHY ONE BETTER THAN OTHER**

---



ONE HOT ENCODING is better because the problem in label here is since there are different numbers in the same column, the model will misunderstand the data to be in some kind of order, 0 < 1 <2.

The model may derive a correlation like as the country number increases the population increases but this clearly may not be the scenario in some other data or the prediction set. To overcome this problem, we use One Hot Encoder.

## **Machine Learning Model**

````python
important_features = ['MarriedID', 'MaritalStatusID', 'SEXID','Termd',
       'EmpStatusID', 'DeptID', 'PerfScoreID', 'FromDiversityJobFairID',
       'Salary', 'PositionID', 'Position', 'State', 'CITIZEND', 'HispanicLatino', 'RaceDesc',
       'TermReason', 'RecruitmentSource',
       'PerformanceScore', 'EngagementSurvey', 'EmpSatisfaction',
       'SpecialProjectsCount', 'DaysLateLast30', 'Absences', 'family_members',
       'ESPS']
````

````python
df_model = df[important_features]
````

````python
categorical_cols = list(df_model.select_dtypes(include = ['object']))
categorical_cols
````

````python
from sklearn.preprocessing import LabelEncoder
df_model[categorical_cols] = df_model[categorical_cols].apply(LabelEncoder().fit_transform)
````

````python
df_model.head(20)
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X = df_model.drop('Termd', axis = 1)
y = df_model['Termd']
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 2)
````

````python
y_train.unique()
````

### **Logistic Regression**

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

model = LogisticRegression()
model.fit(X_train, y_train)

prediction = model.predict(X_test)

print("Accuracy = ", accuracy_score(y_test, prediction))
print("Precision = ", precision_score(y_test, prediction))
print("Recall = ", recall_score(y_test, prediction))
print("F1 Score = ", f1_score(y_test, prediction))
````

### **KNeighbors Classifier**

````python
from sklearn.neighbors import KNeighborsClassifier
classifier= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )
classifier.fit(X_train, y_train)
prediction2 = classifier.predict(X_test)
````

````python
print("Accuracy = ", accuracy_score(y_test, prediction2))
print("Precision = ", precision_score(y_test, prediction2))
print("Recall = ", recall_score(y_test, prediction2))
print("F1 Score = ", f1_score(y_test, prediction2))
````
